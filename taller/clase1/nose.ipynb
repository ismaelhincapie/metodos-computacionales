```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
import seaborn as sns
```

```python
def catalan(n):
    cn=1
    if n == 0:
        return 1
    else:
        for i in range(1, n):
            cn= ((4*i - 2) * cn) // (i + 1)
            i=i+1
        return cn


# Calcular C_50
c50 = catalan(50)
print("C_50 =", c50)

def mcd(m, n):
    if n == 0:
        return m
    else:
        return mcd(n, m % n)


# Calcular MCD de 108 y 192
resultado_mcd = mcd(108, 192)
print("MCD(108, 192) =", resultado_mcd)
```

```python
# Parámetros del problema
v0 = 4.0                # velocidad inicial (m/s)
theta = np.radians(45) # convertir a radianes
g = 9.8                # gravedad (m/s^2)
y0 = 0                 # altura inicial
# Componentes de la velocidad
v0x = v0 * np.cos(theta)
v0y = v0 * np.sin(theta)

# Tiempo total de vuelo
t_final = (2 * v0y) / g

# vector de tiempo
t = np.linspace(0, t_final, 100)

# Ecuaciones del movimiento

x = v0x * t
y = y0 + v0y * t - 0.5 * g * t**2

# Gráfica
plt.figure()
plt.plot(x, y)

plt.xlabel("Distancia horizontal (m)")
plt.ylabel("Altura (m)")
plt.title("Trayectoria de un proyectil (v0 = 4 m/s, θ = 45°)")

plt.grid()
plt.show()
```

```python
# Constantes
G = 6.67430e-11  # Constante de gravitación universal (m^3 kg^-1 s^-2)
M = 5.972e24     # Masa de la Tierra (kg)

def propiedades_orbita(l1, v1):

    # Coeficientes de la ecuación cuadrática
    A = 1
    B = -(2 * G * M) / (v1 * l1)
    C = -(v1**2 - (2 * G * M) / l1)

    # Discriminante
    D = B**2 - 4*A*C

    # Raíces
    v2_1 = (-B + math.sqrt(D)) / (2*A)
    v2_2 = (-B - math.sqrt(D)) / (2*A)

    # Tomamos la raíz positiva más pequeña
    v2 = min(v for v in [v2_1, v2_2] if v > 0)

    # Distancia en el afelio
    l2 = l1 * (v1 / v2)

    # Parámetros de la elipse
    a = 0.5 * (l1 + l2)
    b = math.sqrt(l1 * l2)
    e = (l2 - l1) / (l2 + l1)

    # Periodo orbital
    T = (2 * math.pi * a * b) / (l1 * v1)

    return l2, v2, T, e


# Datos de la Tierra
l1_user = 1.4710e11
v1_user = 3.0287e4

l2, v2, T, e = propiedades_orbita(l1_user, v1_user)

print("\nResultados:")
print(f"Distancia en el afelio (l2) = {l2:.5e} m")
print(f"Velocidad en el afelio (v2) = {v2:.5e} m/s")
print(f"Periodo orbital (T) = {T:.5e} s")
print(f"Excentricidad (e) = {e:.5f}")
```
