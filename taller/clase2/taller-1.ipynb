```python
def f(v):
    return 8 - v - 3* np.exp(0.5*v)

V = np.linspace(1,2,100)
y = f(V)

plt.plot(V, y)
plt.xlabel('v')
plt.ylabel('f(v)')
plt.title('Grafica de f(v) = 8 - v - 3 * e^(0.5 * v)')
plt.grid(True)
plt.show()
```

```python
# función
def f(v):
    return 8 - v - 3*np.exp(0.5*v)

# método de bisección
def biseccion(f, a, b, tol):

    if f(a)*f(b) >= 0:
        print("El intervalo no es válido")
        return None

    error = 100
    p_old = a
    i = 0
    print("i        a        b        p        f(p)      error(%)")

    while error > tol:

        p = (a+b)/2

        if i > 0:
            error = abs((p - p_old)/p)*100

#        print("Iteración:", i)
        print(f"{i:<3} {a:.6f} {b:.6f} {p:.6f} {f(p):.6f} {error:.6f}")
        print()


        if f(a)*f(p) < 0:
            b = p
        else:
            a = p

        p_old = p
        i += 1

    return p

# resolver entre 1 y 2 con una tolerancia del 2%
raiz = biseccion(f,1,2,2)

print("Raíz aproximada:",raiz)


# gráfica
V = np.linspace(0,3,200)
y = f(V)

plt.plot(V,y,label="f(V)")
plt.axhline(0)

plt.scatter(raiz,f(raiz),color='red',label="Raíz")

plt.xlabel("Voltaje (V)")
plt.ylabel("f(V)")
plt.title("Método de Bisección")
plt.legend()
plt.grid()

plt.show()
```
