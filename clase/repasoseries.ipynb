```python
def taylor_polynomial(f, x0, n, h=1e-5):
    """
    Compute Taylor polynomial coefficients for function f around point x0.
    
    Parameters:
    f: function to approximate
    x0: center point for Taylor expansion
    n: number of terms (order + 1)
    h: small value for numerical derivative approximation
    
    Returns:
    coefficients: list of Taylor coefficients
    """
    coefficients = []
    for k in range(n):
        # Numerical derivative approximation
        def derivative(func, point, order, step):
            if order == 0:
                return func(point)
            return (derivative(func, point + step, order - 1, step) - 
                    derivative(func, point - step, order - 1, step)) / (2 * step)
        
        coef = derivative(f, x0, k, h) / np.math.factorial(k)
        coefficients.append(coef)
    
    return coefficients

# Example: Taylor polynomial for sin(x) around x=0
taylor_coeffs = taylor_polynomial(np.sin, 0, 5)
print("Taylor coefficients for sin(x):", taylor_coeffs)

# Evaluate Taylor polynomial at x=0.2
result = sum(coef * x**k for k, coef in enumerate(taylor_coeffs))
print(f"Taylor approximation at x={x}: {result}")
print(f"Actual sin({x}): {np.sin(x)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# dominio
x = np.linspace(-2,2,400)

# función real
f = np.arctan(x)

# polinomios de Taylor
T1 = x
T3 = x - x**3/3
T5 = x - x**3/3 + x**5/5
T7 = x - x**3/3 + x**5/5 - x**7/7
T9 = x - x**3/3 + x**5/5 - x**7/7 + x**9/9

plt.figure(figsize=(8,5))

plt.plot(x,f,label='arctan(x)',linewidth=3)
plt.plot(x,T1,label='Orden 1')
plt.plot(x,T3,label='Orden 3')
plt.plot(x,T5,label='Orden 5')
plt.plot(x,T7,label='Orden 7')
plt.plot(x,T9,label='Orden 9')

plt.legend()
plt.grid()

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Serie de Taylor de arctan(x)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#funcion original
def f(x):
    return 1 / (1 + x)

#expansion de taylor hasta el orden N
def taylor_1_over_1px(x, N):
    taylor_sum = np.zeros_like(x)#creamos un array de ceros del mismo tamaño que x
    for n in range(N+1): #iteramos desde 0 hasta N
        taylor_sum += ((-1)**n) * x**n #sumamos cada término de la serie de Taylor
    return taylor_sum

# rango de x
x = np.linspace(-0.9, 2, 400)
y = f(x)

# ordenes de la expansion
ordenes = [1, 2, 3, 5, 10]

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x)=1/(1+x)', linewidth=2, color='black')

for N in ordenes:
    y_taylor = taylor_1_over_1px(x, N)
    plt.plot(x, y_taylor, label=f'Taylor orden {N}')

plt.axvline(0, color='gray', lw=0.5, ls='dashed')
plt.ylim(-2, 2)
plt.xlim(-0.9, 2)
plt.legend()
plt.title("Expansión de Taylor de $f(x)=1/(1+x)$ alrededor de $x=0$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

print("f(0.02) = ",f(0.02))

```
