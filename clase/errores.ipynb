```python
#el error es determinado por el intrumento de medida y el procedimiento de medicion, 
#por lo tanto es necesario conocer el error para poder determinar la incertidumbre de la medicion.

#para no afectar la medida por la reaccion del ser humano se hacen muchas mediciones y se promedia el resultado, 
#esto se llama error aleatorio, el error sistematico es aquel que se repite en todas las mediciones y es causado 
#por un error en el instrumento de medida o en el procedimiento de medicion.

```

```python
def calcular_errores(valor_medido, valor_teorico):
    """
    Calcula el error relativo y porcentual.
    
    Args:
        valor_medido: valor obtenido en la medición
        valor_teorico: valor teórico o valor de referencia
    
    Returns:
        diccionario con error absoluto, relativo y porcentual
    """
    error_absoluto = abs(valor_medido - valor_teorico)
    error_relativo = error_absoluto / abs(valor_teorico)
    error_porcentual = error_relativo * 100
    
    return {
        'error_absoluto': error_absoluto,
        'error_relativo': error_relativo,
        'error_porcentual': error_porcentual
    }
valor_medido = float(input("Ingrese el valor medido: "))
valor_teorico = float(input("Ingrese el valor teórico: "))
valores = calcular_errores(valor_medido, valor_teorico)

print("Error absoluto:", valores['error_absoluto'])
print("Error relativo:", valores['error_relativo'])
print("Error porcentual:", valores['error_porcentual'])
```
